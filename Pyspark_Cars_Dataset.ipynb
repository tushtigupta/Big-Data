```python
df=spark.read.option("header",True).csv("/content/cars.csv")
```

```python
#Check the summary
df.describe().show()
```

```python
#select
df.select(['brand','fuel']).show()
```

```python
#add
df_revised = df.withColumn("Discounted Price", df['selling_price']-(df['selling_price']/100))
df_revised.show()
```

```python
#drop
df_final_columns = df_revised.drop('Discounted Price')
df_final_columns.show()
```

```python
# Rename the column
df_final = df_final_columns.withColumnRenamed("selling_price", "Selling_price")
df_final.show(3)
```

```python
df_final.createOrReplaceTempView("cars_data")
df_sql = spark.sql("select * from cars_data")
df_sql.show()
```
