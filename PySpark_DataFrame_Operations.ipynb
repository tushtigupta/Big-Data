```python
# Create DataFrame
df = spark.createDataFrame(data, columns)
```

```python
# Show DataFrame
df.show(truncate=False)
```

#OPERATIONS

```python
#1--Filter employees with salary greater than 50,000.
df.filter(df.salary > 50000).show()
```

```python
#2--Filter employees whose age is above 30
df.filter(df.age > 30).show()
```

```python
#3--Count total number of employees.
df.count()
```

```python
#4--Show all employees from a specific city (e.g., "Delhi").
df.filter(df.city == "Delhi").show()
```

```python
#5--Select only name and salary columns
df.select("name", "salary").show()
```

#GroupBy() OPERATIONS

```python
#6--Group by department and count employees
df.groupBy("department").count().show()
```

```python
#7--Find total salary per department
from pyspark.sql.functions import sum

df.groupBy("department").agg(sum("salary").alias("total_salary")).show()
```

```python
#8--Find average salary per department
from pyspark.sql.functions import avg

df.groupBy("department").agg(avg("salary").alias("avg_salary")).show()
```

```python
#9--Group by city and find max salary
from pyspark.sql.functions import max

df.groupBy("city").agg(max("salary").alias("max_salary")).show()
```

```python
#10--Group by department and gender and count employees
df.groupBy("department", "gender").count().show()
```

#Multiple Aggregation

```python
#11_&_12--Multiple aggregations on salary per department
from pyspark.sql.functions import sum, avg, max, min

df.groupBy("department").agg(
    sum("salary").alias("total_salary"),
    avg("salary").alias("avg_salary"),
    max("salary").alias("max_salary"),
    min("salary").alias("min_salary")
).show()
```

```python
#13--Group by department and role
from pyspark.sql.functions import count

df.groupBy("department", "role").agg(
    sum("salary").alias("total_salary"),
    avg("salary").alias("avg_salary"),
    count("*").alias("employee_count")
).show()
```

```python
#14--Count distinct roles in each department
from pyspark.sql.functions import countDistinct

df.groupBy("department").agg(
    countDistinct("role").alias("distinct_roles")
).show()
```

```python
#15--Top 3 departments with highest total salary
df.groupBy("department").agg(
    sum("salary").alias("total_salary")
).orderBy("total_salary", ascending=False).limit(3).show()
```
